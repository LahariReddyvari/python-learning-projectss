```python
# Ask the user to choose:rock, paper, scissors
# The computer randomly chooses one too. Print both choices.
# Decide who wins.
# Ask:Play again? (y/n)
import random 
ask = "y" #initializing the condition before the loop.
while ask.lower() == "y": 
    game = input("choose:rock, paper, scissors?").lower() 
    computer_chose = random.choice(["rock", "paper", "scissors"]).lower()
    print("You chose:", game)
    print("Computer chose:", computer_chose) 

    win1 = game == "rock" and computer_chose == "scissors" 
    win2 = game == "paper" and computer_chose == "rock" 
    win3 = game == "scissors" and computer_chose == "paper" 

    if game == computer_chose: 
        print("It's a tie!") 
    elif win1 or win2 or win3: 
        print("You win!") 
    else: 
        print("Computer wins!") 
    ask = input("Play again? (y/n): ")
print("Thanks for playing! 😊")
```

# Rock Paper Scissors Game
A simple Python game where the user plays Rock, Paper, Scissors against the computer.
The computer randomly selects one option, and the program compares both choices to determine the winner.

## Game Rules:
- Rock beats Scissors
- Scissors beats Paper
- Paper beats Rock
- If both choices are the same, the result is a tie

## Concepts Learned:
- Importing modules (`random`)
- Using `random.choice()` for random selection
- While loops for repeated gameplay
- If/elif/else conditional statements
- Boolean operators (`and`, `or`)
- User input handling
- String methods (`.lower()`)

## How It Works:
1. The user selects rock, paper, or scissors.
2. The computer randomly selects a choice.
3. The program compares both selections.
4. The winner is displayed.
5. The user can choose to play again.
