```python
#Example 1
# Ask the user:Toss the coin? (y/n):If the user enters y or Y:Randomly print either:Heads or Tails
# Keep asking until the user enters n.
# If the user enters anything else, print:Invalid choice!

import random
coin = input("Toss the coin? (y/n): ")
# while coin.lower() == "y": this converts the Y to small y.

while coin.lower() == "y":
    result = random.choice(["Heads", "Tails"])
    print(result)
    coin = input("Toss the coin? (y/n): ")
if coin.lower() == "n":
    print("Thanks for playing!")
else:
    print ("Invalid Choice!")
```

# Coin Toss Game
A simple Python game where the user can repeatedly toss a coin.
The computer randomly generates either Heads or Tails.
