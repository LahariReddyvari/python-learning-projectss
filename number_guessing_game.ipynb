````python
# The computer randomly picks a number between 1 and 10. Ask the user:Guess the number:
# If they guess correctly:Congratulations! You guessed it.
# Otherwise:Try again!
# Keep asking until they guess correctly.
import random
number = int(input("Guess the number between 1/10:"))
guess_number = random.randint(1, 10)
while number != guess_number:
    print("Try again!")
    number = int(input("Guess the number between 1/10:"))
print("Congratulations! You guessed it.")
````

# Number Guessing Game 

A simple Python game where the computer randomly selects a number between 1 and 10, and the user keeps guessing until they find the correct number.

## Concepts Learned:
- Importing modules (`random`)
- Generating random numbers using `random.randint()`
- User input using `input()`
- Type conversion using `int()`
- While loops
- Comparison operators
